# Representations across the same model

```python
import seaborn as sns
import matplotlib.pyplot as plt
import json
from pathlib import Path
```

```python
file_path = Path('cka_bert-base-uncased_bert-base-uncased.json')

with file_path.open('r', encoding='utf-8') as file:
    bert = json.load(file)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(bert, fmt=".1f", cmap='viridis', cbar_kws={'label': 'CKA (Linear)'})
plt.xlabel(f"Capas de bert-base-uncased")
plt.ylabel(f"Capas de bert-base-uncased")
plt.ylim(0, len(bert))
plt.xticks(ticks=[i + 0.5 for i in range(len(bert))], labels=range(1, len(bert) + 1))
plt.yticks(ticks=[i + 0.5 for i in range(len(bert))], labels=range(1, len(bert) + 1))
plt.show()
```

```python
plt.figure(figsize=(10, 8))

x = range(1, len(bert[0]) + 1) 
for i, row in enumerate(bert):
    plt.plot(x, row, label=f'Capa {i + 1}')  # Plot each row with a label

plt.xlabel("Capas bert-base-uncased")
plt.ylabel("CKA (Linear)")
plt.title("CKA (Linear) para cada Capa de bert-base-uncased")
plt.legend(title="Capa")  # Add a legend with a title
plt.grid(True)  # Optionally add a grid for better readability
plt.xticks(range(1,13))
plt.show()
```

```python
file_path = Path('cka_wav2vec2_wav2vec2.json')

with file_path.open('r', encoding='utf-8') as file:
    w2v = json.load(file)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(w2v, fmt=".1f", cmap='viridis', cbar_kws={'label': 'CKA (Linear)'})
plt.xlabel(f"Capas de Wav2Vec 2.0")
plt.ylabel(f"Capas de Wav2Vec 2.0")
plt.ylim(0, len(w2v))
plt.xticks(ticks=[i + 0.5 for i in range(len(w2v))], labels=range(1, len(w2v) + 1))
plt.yticks(ticks=[i + 0.5 for i in range(len(w2v))], labels=range(1, len(w2v) + 1))
plt.show()
```

```python
colors = [
    '#e6194B',  # bright red
    '#3cb44b',  # bright green
    '#ffe119',  # bright yellow
    '#4363d8',  # bright blue
    '#f58231',  # bright orange
    '#911eb4',  # purple
    '#46f0f0',  # turquoise
    '#f032e6',  # magenta
    '#bcf60c',  # lime green
    '#fabebe',  # pink
    '#008080',  # teal
    '#e6beff'   # lavender
]

plt.figure(figsize=(12, 8))

x = range(1, len(w2v[0]) + 1)
for i, row in enumerate(w2v):
    plt.plot(x, row, label=f'Capa {i + 1}', color=colors[i])
    plt.scatter(i + 1, row[i], color='black')  # Marca en la diagonal (identidad)
    plt.scatter(7, row[6], color='red')  # Marca cuando x = 7

plt.xlabel("Capas de Wav2Vec 2.0", fontsize=15)
plt.ylabel("Linear CKA", fontsize=15)
plt.legend(title="Capa", bbox_to_anchor=(1.05, 1), loc='upper left')  # Mueve la leyenda fuera del grfico
plt.xticks(range(1, 13))
plt.show()
```
