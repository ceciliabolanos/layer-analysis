```python
import json 
import pandas as pd
from collections import Counter

with open('retrieval_base_layer7_bert-base-uncased_layer1_p4_k100.json', 'r') as f:
    encodecmae_bert = json.load(f)

with open('retrieval_mel256-ec-base_layer7_bert-base-uncased_layer0_p4_k100.json', 'r') as f:
    encodecmae_mel_bert = json.load(f)

with open('retrieval_wav2vec2_layer6_bert-base-uncased_layer0_p4_k100.json', 'r') as f:
    w2v_bert = json.load(f)
```

# Incorrectly returned words

We consider words to be incorrectly returned if the word that should appear is not in the top 5.

```python
keys_test = [encodecmae_bert['all_values'][i] for i in encodecmae_bert['rows_deleted']]
total = len(keys_test)
words_encodec = []
count = 0
for i in range(total):
    word = keys_test[i]
    k_1 = encodecmae_bert['retrieval_indices'][i]
    if not any(encodecmae_bert['keys_anchors'][index] == word for index in k_1 if index < len(encodecmae_bert['keys_anchors'])):
        word_retrieval =  [encodecmae_bert['keys_anchors'][k_1] for k_1 in encodecmae_bert['retrieval_indices'][i]]
        words_encodec.append([word, word_retrieval])
        count +=1 
        print(f"word: {word} retrieval:{word_retrieval}, indice {i}")    
        print('---------')
print(f'porcentaje de no devueltas: {count/total}')        
print(f'palabras: {count}') 
```

```python
keys_test = [encodecmae_mel_bert['all_values'][i] for i in encodecmae_mel_bert['rows_deleted']]
total = len(keys_test)
words_encodec_mel = []
count = 0
for i in range(total):
    word = keys_test[i]
    k_1 = encodecmae_mel_bert['retrieval_indices'][i]
    if not any(encodecmae_mel_bert['keys_anchors'][index] == word for index in k_1 if index < len(encodecmae_mel_bert['keys_anchors'])):
        word_retrieval =  [encodecmae_mel_bert['keys_anchors'][k_1] for k_1 in encodecmae_mel_bert['retrieval_indices'][i]]
        words_encodec_mel.append([word, word_retrieval])
        count +=1 
        print(f"word: {word} retrieval:{word_retrieval}, indice {i}")    
        print('---------')
print(f'porcentaje de no devueltas: {count/total}')        
print(f'palabras: {count}') 
```

```python
keys_test = [w2v_bert['all_values'][i] for i in w2v_bert['rows_deleted']]
total = len(keys_test)
words_w2v = []
count = 0
for i in range(total):
    word = keys_test[i]
    k_1 = w2v_bert['retrieval_indices'][i]
    if not any(w2v_bert['keys_anchors'][index] == word for index in k_1 if index < len(w2v_bert['keys_anchors'])):
        word_retrieval =  [w2v_bert['keys_anchors'][k_1] for k_1 in w2v_bert['retrieval_indices'][i]]
        words_w2v.append([word, word_retrieval])
        count +=1 
        print(f"word: {word} retrieval:{word_retrieval}, indice {i}")    
        print('---------')
print(f'porcentaje de no devueltas: {count/total}')              
print(f'palabras: {count}') 
```
