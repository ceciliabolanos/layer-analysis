```python
import json
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
file_path_heatmap = Path('cka_wav2vec2_encodecmae_base.json')

with file_path_heatmap.open('r', encoding='utf-8') as file:
    encodec_w2v = json.load(file)

plt.figure(figsize=(8, 6))
sns.heatmap(encodec_w2v, fmt=".1f", cmap='viridis', cbar_kws={'label': 'CKA (Linear)'})
plt.xlabel(f"Layer wav2vec2")
plt.ylabel(f"Layer Encodecmae")
plt.ylim(0, len(encodec_w2v))
plt.show()

```
